```
root_path = 'gdrive/My Drive/data'
```

```
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import os
import json
```

```
domain_dic = {}
domain_list_wifi = []
domain_list_lte = []
domain_dic['wifi'] = {}
domain_dic['lte'] = {}

# print (os.listdir(root_path)[0])
for folder in os.listdir(root_path):
  print (folder)
  for file in os.listdir(root_path+'/'+folder+'/json'):
    with open(root_path+'/'+folder+'/json/'+file) as f:
      data = json.load(f)
      url_name = data[0]["baseDomain"]
      print (url_name)
      if 'WIFI' in folder:        
        domain_dic["wifi"][url_name] = data[0]
        domain_list_wifi.append(url_name)
      else:
        domain_dic["lte"][url_name] = data[0]
        domain_list_lte.append(url_name)
      f.close()
      
domain_list = [f for f in domain_list_wifi if f in domain_list_lte]
```

```
print ("First Party Requests")
# print (len(domain_list))
for domain in domain_list:
  print (domain)
  print ("wifi",domain_dic['wifi'][domain]['firstParty']['requests'])
  print ("lte",domain_dic['lte'][domain]['firstParty']['requests'], '\n')  
```

```
print ("Third Party Requests")
# print (len(domain_list))
for domain in domain_list:
  print ("domain:",domain)
  try:
    print ("wifi:",domain_dic['wifi'][domain]['thirdParty']['requests'])
  except:
    print ("wifi: json file exception occured")
   
  try:
    print ("lte:",domain_dic['lte'][domain]['thirdParty']['requests'], '\n')
  except:
    print ("lte: json file exception occured \n")
    
    
   
```

```
firstPartyrequestsLTE = [domain_dic['lte'][domain]["firstParty"]["requests"] for domain in domain_list]
firstPartyrequestsWifi = [domain_dic['wifi'][domain]["firstParty"]["requests"] for domain in domain_list]
# for domain in domain_list:
#   print (domain_dic['lte'][domain]["firstParty"]["requests"])
len(firstPartyrequestsLTE)
```

```


firstPartyrequestsLTE = tuple(firstPartyrequestsLTE)
firstPartyrequestsWifi = tuple(firstPartyrequestsWifi)

ind = np.arange(len(firstPartyrequestsLTE))  # the x locations for the groups
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(ind - width/2, firstPartyrequestsLTE, width,
                label='LTE')
rects2 = ax.bar(ind + width/2, firstPartyrequestsWifi, width,
                label='Wifi')

labels = [domain for domain in domain_list]

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('First Party Requests')
ax.set_title('First Party Requests (LTE vs Wifi)')
ax.set_xticks(ind)
ax.set_xticklabels(tuple(labels))
ax.legend()


def autolabel(rects, xpos='center'):
    """
    Attach a text label above each bar in *rects*, displaying its height.

    *xpos* indicates which side to place the text w.r.t. the center of
    the bar. It can be one of the following {'center', 'right', 'left'}.
    """

    ha = {'center': 'center', 'right': 'left', 'left': 'right'}
    offset = {'center': 0, 'right': 1, 'left': -1}

    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(offset[xpos]*3, 3),  # use 3 points offset
                    textcoords="offset points",  # in both directions
                    ha=ha[xpos], va='bottom')


autolabel(rects1, "left")
autolabel(rects2, "right")

fig.tight_layout()
plt.rcParams["figure.figsize"] = [70,9]
# plt.figure(figsize=(10,9))
plt.savefig('fig1.png', dpi = 300)

# plt.show()
```
